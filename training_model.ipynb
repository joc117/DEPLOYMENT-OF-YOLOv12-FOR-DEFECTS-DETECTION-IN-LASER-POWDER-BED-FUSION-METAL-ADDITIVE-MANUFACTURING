```python
from ultralytics import YOLO

model = YOLO('runs/detect/yolov12-defect-model/weights/best.pt')
results = model.val(data= f'{DATASET_PATH}/data.yaml')

# results.box has per-class metrics in a dict
for i, class_name in enumerate(results.names.values()):
    p = results.box.p[i]
    r = results.box.r[i]
    f1 = 2 * p * r / (p + r) if (p + r) > 0 else 0
    print(f"{class_name}: P={p:.3f}, R={r:.3f}, F1={f1:.3f}")
```
